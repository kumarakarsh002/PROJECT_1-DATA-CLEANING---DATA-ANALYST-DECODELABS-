```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Dataset for Data Analytics.csv")
df
```

```python
print("Initial Dataset Info:")
df.info()
```

```python
print("\n--- Missing Values Count ---")
print(df.isnull().sum())
```

```python
total_duplicates = df.duplicated().sum()
print(f"Total duplicate rows: {total_duplicates}")
```

```python
duplicate_orders = df.duplicated(subset=['OrderID']).sum()
print(f"Duplicate Order IDs: {duplicate_orders}")
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop_duplicates(subset=['OrderID'])

print("Duplicates handled successfully!")
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
print(df[['Date']].info())
```

```python
df['CouponCode'] = df['CouponCode'].fillna('No Coupon')
print(df.isnull().sum())
```

```python
df.to_csv('Cleaned_Data_Analytics.csv', index=False)
print("Data cleaning complete! File saved as 'Cleaned_Data_Analytics.csv'")
```
